```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Melakukan export file
df= pd.read_excel('gpv_model.xlsx',  sheet_name='input')
```

```python
#memanggil file bernama df
df
```

```python
df = df.astype(int)
```

```python
df.info()
```

```python
# membuat kolom baru bernama "cf_res_g" dan memanggil fungsi diff()
df['cf_res_g'] = df['gross_reserve'].diff(-1)
```

```python
df
```

```python
# membuat kolom baru bernama "cf_res_n" dan memanggil fungsi diff()
df['cf_res_n'] = df['net_reserve'].diff(-1)
```

```python
df
```

```python
# membuat kolom "col_g" , "col_n" , dan "expense_t" 
df['col_g'] = df['gross_claim_paid'] + df['gross_lapse_cost']
df['col_n'] = df['net_claim_paid'] + df['net_lapse_cost']
df['expense_t'] = df['expense'] + df['xol_cost']
```
